## Multiplying matrices and vectors

** Matrices Product:** The matrix product of matrices $\textbf{A}$ and $\textbf{B}$ is a third matrix $\textbf{C}$.

** Properties of Matrices products: ** 
* *Distributive:* $A(B + C) =AB + AC$
* *Associative:* $ABC = (AB)C$
* The transpose of a matrix product has a simple form $(AB)^T = B^TA$

**Note**: Matrix maltiplication is not commutative i.e $AB \neq BA$

** Hadamard product (element-wise product): ** a matrix product containing the product of the individual elements denoted as $A\odot B$.

** Dot product ** between two vectors $x$ and $y$ of the same dimensionality is the matrix product $\mathbf{x^Ty}$. The dot product between two vectors is commutative i.e $$\mathbf{x^T y=y^Tx}$$.



```python
# Let create a matrix A and B
A = np.matrix([[ 1.,  0., 1.],
        [ -1.,  1., 0.],
        [1.,  0.,  -1.]])

B = np.matrix([[2,1,-2],
              [-2,2,1],
              [1,-2,2]])
```

```python
# Matrix product:
C =  A* B
C
```

```python
C2 = np.dot(A,B)
C2
```

It clear that the product of two matrix() is equal to their dot product. However, the product of two array()  is not equal to the matrix product. The product of two array()  is equal to Hadamard product.

```python
# # Let create an array of  A and B
A1 = np.array([[ 1.,  0., 1.],
        [ -1.,  1., 0.],
        [1.,  0.,  -1.]])

B1 = np.array([[2,1,-2],
              [-2,2,1],
              [1,-2,2]])
D = A1*B1
D
```

It clear that D is not equal to C. To obtain matrix product from two array we use the dot() function

```python
#Use dot product
D = np.dot(A1,B1)
D
```

Now D is equal to C

## Identity and Inverse Matrices

An *identity matrix* is a matrix that does not change any vector when we multiply that vector by that matrix denoted as $I_n \in  \mathbb{R}^{n \times n}$.

The matrix inverse of $A$ is denoted as $A^{-1}$, is defined as the matrix such that:
$$ \mathbf{A A^{-1} = I_n} $$

```python
# The inverse of matrix A
from scipy.linalg import inv

# The inverse of matrix A is 
print ("The Inverse of Matrix  A = ", A, "is   ", inv(A))
```

## Finding Determinant

The determinant of a square matrix $\mathbf{A}$ is often denoted $\mid\mathbf{A}\mid$ and is a quantity often used in linear algebra. In SciPy this is computed as shown in this example:

```python
from scipy import linalg
print ("The Determinant of Matrix  A = ", A, "is   ", linalg.det(A))
```

## Linear Equations

A system of linear equations is given as $$ \mathbf{Ax =b}$$ where $A \in \mathbb{R}^{m \times n}$ is a known matrix, $b \in \mathbb{R}^{m}$ is a known vector, and $x \in \mathbb{R}^n$ is a
vector of unknown variables. We can solve for $\mathbf{x}$ by following steps:

\begin{align*}
Ax &=b \\
A^{-1} Ax &= A^{-1} b \\
I_n x &=A^{-1} b \\
x &=A^{-1} b
\end{align*}

For example, let solve these equations: \begin{eqnarray*} x + 3y + 5z & = & 10 \\
                   2x + 5y + z & = & 8  \\
                   2x + 3y + 8z & = & 3
 \end{eqnarray*}

```python
# In matrix notation: 
A = np.matrix([[1., 3., 5.],
               [2., 5., 1.],
               [2., 3., 8.]])

b = np.array([10,8,3])

#using a matrix inverse

x = inv(A).dot(b)
print(x)

```

```python
# Or we may solve using scipy solve command which can be faster and more numerically stable
from scipy.linalg import solve

x = solve(A,b)
print(x)
```

```python
# Alternatively we can use numpy slove
x = np.linalg.solve(A, b)
print(x)
```

## Norms

* **Norms** measure the size of the vectors and is given by: $$ \Vert x \Vert _p = \Big( \sum_i  \mid x_i \mid ^p\Big)^{1/p} $$ for
$p \in \mathbb{R}, p \leq 1$.

* **Euclidean Distance $ L^2$** is the norm with $p=2$ which is simply the distance from the origin to the point indenified by $x$ given by: $$ \Vert x \Vert _2 = \Big( \sum_i  \mid x_i \mid ^2\Big)^{1/2} $$.

* **The Squared $ L^2$ Norm** is the size of the vector calculated simply as $x^Tx$. This is more convinient to work with mathematically and computationally.


* **The $L^1$ Norm** may be simplified to $$ \Vert x \Vert _1 = \Big( \sum_i  \mid x_i \mid \Big) $$.

* **The $L^\infty$ Norm** also knowas the max norm is the value of the element with largest magnitude in vector.
$$\Vert x \Vert _\infty = \max_{i} \mid x_i \mid$$.

* **Frobenius Norm** this measure the size of a matrix and is anologous to the $L^2$ norm of a vector given by:
$$\Vert A \Vert _F = \sqrt{\sum_{i,j} A_{i,j}^2}  $$


#### We can use numpy or scipy liag.norm function to find the norms of a vector or matrix.

The syntax is the same: *scipy.linalg.norm(a, ord=None, axis=None, keepdims=False)* more details here for [scipy](https://docs.scipy.org/doc/scipy-0.17.0/reference/generated/scipy.linalg.norm.html) and [numpy](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.norm.html#numpy.linalg.norm).

```python
from scipy.linalg import norm
## The Frobenius norm of Matrix A
A = np.matrix([[1., 3., 5.],
               [2., 5., 1.],
               [2., 3., 8.]])
print ("The Frobenius norm of matrix A= ", A ,"is ", norm(A,'fro')) # frobenius norm is the default
```

```python
##Euclidean Distance 
b = np.array([1,3,6,9,12])
print ("The Euclidean Distance of vector b= ", b ,"is ", np.linalg.norm(b))
```

```python
## The Squared Norm

print( "The Squared Norm of vector b= ", b ,"is ", b.T.dot(b))
```

```python
# The  L1  Norm
print ("The L1 Norm of vector b= ", b ,"is ", norm(b,1)) # L1 norm (max column sum)
```

```python
# the max norm
print ("The Max Norm of vector b= ", b ,"is ", norm(b, np.inf))
```
